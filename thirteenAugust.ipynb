```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread(r"C:/Users/akhil/Documents/dip/img9.tif",cv2.IMREAD_GRAYSCALE)
cv2.imshow('MRI',image)
cv2.waitKey(10000)
cv2.destroyAllWindows()
image.shape #((452, 374), # in RGB,3)
image.size #169048*3
image.dtype #uint8
```

```python
bit = [0,1,2,3,4,5,6,7,8]

for i in range(0,8):
    levels = 2**bit[i+1]
    factor = 256//levels
    image_lowbit = (image//factor)*factor
    plt.subplot(2,4, i+1)
    plt.imshow(image_lowbit, cmap='gray')
```
